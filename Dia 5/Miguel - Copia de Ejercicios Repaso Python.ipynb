## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
contador = 0
#Respuesta:
#Debe dar 5

for num in nums:
    contador += 1
    if num == 0:
        break


print(contador)
```

## Ejercicio 3

Calcular el valor medio de los valores que hay en una lista de enteros. Calcular solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
contador = 0
#Respuesta:
#Debe dar 4.2
for numero in nums:
    suma += numero
    contador += 1
    if numero == 0:
      break

valor_medio = suma / contador
print(valor_medio)
```

## Ejercicio 5 (Opcional)

Crear una lista a partir de una secuencia de valores de entrada terminada en *.
Una vez creada, modificarla elevando al cuadrado todos los pares y al cubo todos los impares, y mostrarla.

```python
lista = [3,56,6,7,"*",2,78]
nueva_lista = []
for elemento in lista:
  # print (elemento)
  if elemento == '*':
    break
  nueva_lista.append(elemento)
# print(nueva_lista)
lista_resultado=[]
for numero in nueva_lista:
  if numero %2 == 0:
    lista_resultado.append(numero**2)
  elif numero %2 != 0:
    lista_resultado.append(numero**3)
print(lista_resultado)
#Respuesta:
#Debe dar [27, 3136, 36, 343]
```
